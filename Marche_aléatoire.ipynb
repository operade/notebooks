marche aleatoire dans $\mathbb Z^2$




   




```
from numpy.random import choice
Liste = [[0,1],[0,-1],[1,0],[-1,0]]
def marche(n):
  indices = [0,1,2,3]
  S = [0,0]
  L = [S]
  for i in range(n):
    i = choice(indices)
    a = S[0]+Liste[i][0]
    b = S[1]+Liste[i][1]
    S = [a,b]
    L.append(S)
  return L
```

```
print(marche(100))
```

```
import matplotlib.pyplot as plt
L = marche(10000)
x = [L[i][0] for i in range(len(L))]
y = [L[i][1] for i in range(len(L))]
plt.plot(x,y)
plt.show()
```
